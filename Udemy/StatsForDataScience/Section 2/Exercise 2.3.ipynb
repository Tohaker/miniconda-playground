```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Categorical Variables: Visualisation Techniques
**Background:** There is an ice cream shop operating in New York, LA, and San Francisco.  
**Data:** You have sold 12,327 ice creams in New York, 17,129 in LA and 19,923 in San Francisco.

```python
# Frequency Distribution Table
index = ['New York', 'LA', 'San Francisco']
data = [12327, 17129, 19923]

table = pd.DataFrame(data, index, columns=['Frequency'])
print(table)
```

```python
# Bar chart
table.plot.bar()
```

```python
# Pie Chart
table.plot.pie(y=0)
```

```python
# Pareto Diagram
table = table.sort_values(by='Frequency', ascending=False)
table['Cumulative Percentage'] = table['Frequency'].cumsum() / table['Frequency'].sum() * 100

fig, ax = plt.subplots()
ax.bar(table.index, table['Frequency'], color='C0')
ax2 = ax.twinx()
ax2.plot(table.index, table['Cumulative Percentage'], color='C1', marker='D', ms=7)

ax.tick_params(axis='y', colors='C0')
ax2.tick_params(axis='y', colors='C1')
plt.show()
```
